```python
# get all the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# get data
df=pd.read_excel(r"C:\Users\Admin\Downloads\Honda.xlsx")
df
```

```python
# get info of data
df.info()
```

```python
# check for null values
df.isnull().sum()
```

```python
# cleaning of data
# delete shipped date column
df.drop(["Shipped Date"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df["Shipper Name"].isna()]
```

```python
df[df.Salesperson=="Nancy Freehafer"]
```

```python
mode=df["Shipper Name"].mode()[0]
mode
```

```python
df["Shipper Name"].fillna(mode,inplace=True)
```

```python
df.isnull().sum()
```

```python
mean=round(df["Unit Price"].mean(),2)
mean
```

```python
median=round(df["Quantity"].median(),0)
median
```

```python
df["Unit Price"].fillna(mean,inplace=True)
```

```python
df["Quantity"].fillna(median,inplace=True)
```

```python
df.isnull().sum()
```

```python
df["month"]=df["Order_Date"].dt.month
df
```

```python
df["month_name"]=df["Order_Date"].dt.month_name()
df
```

```python
df["quarter"]=df["Order_Date"].dt.quarter
df
```

```python
df["sales"]=df["Unit Price"]*df.Quantity
df
```

```python
# region wise sales
rs=df[["Region","sales"]].groupby("Region")[["sales"]].sum()
rs
```

```python
rs.index
```

```python
rs.sales
```

```python
# barplot
# plt.bar(rs.index,rs.sales)
```

```python
# payment wise sales
ps=df[["Payment Type","sales"]].groupby("Payment Type")[["sales"]].sum().sort_values(by="sales",ascending=False)
ps
```

```python
# product  wise sales
prodsales=df[["Product Name","sales"]].groupby("Product Name")[["sales"]].sum().sort_values(by="sales",ascending=False)[:10]
prodsales
```

```python
# category wise sales
catsales=df[["Category","sales"]].groupby("Category")[["sales"]].sum().sort_values(by="sales",ascending=False)[:10]
catsales
```

```python
# city wise sales
citysales=df[["City","sales"]].groupby("City")[["sales"]].sum().sort_values(by="sales",ascending=False)[:10]
citysales
```

```python
# month wise sales
ms=df[["month","sales"]].groupby("month")[["sales"]].sum()
ms
```

```python
# quarter wise sales
qs=df[["quarter","sales"]].groupby("quarter")[["sales"]].sum()
qs
```

```python
# region wise quantity sold
rqs=df[["Region","Quantity"]].groupby("Region")[["Quantity"]].sum()
rqs
```

```python
# total quantity sold
df["Quantity"].sum()
```

```python
# category wise quantity sold
cqs=df[["Category","Quantity"]].groupby("Category")[["Quantity"]].sum()
cqs
```

```python
# shipper name  wise sales
snsales=df[["Shipper Name","sales"]].groupby("Shipper Name")[["sales"]].sum()
snsales
```

```python
# visualisation
fig, ax = plt.subplots(nrows=3, ncols=3,figsize=(14,12))
plt.subplots_adjust(hspace=0.7,wspace=0.8)

ax[0, 0].bar(rs.index,rs.sales,color="g",width=0.5)
ax[0,0].set_ylabel('sales')
ax[0,0].set_xlabel('region')
ax[0,0].set_title('region wise sales')
ax[0, 1].pie(ps.sales,labels=ps.index,autopct="%0.1f%%")

ax[0,1].set_title('payment wise sales')
ax[0, 2].barh(prodsales.index,prodsales.sales,color="g")
ax[0,2].set_xlabel('sales')
ax[0,2].set_ylabel('product')
ax[0,2].set_title('product wise sales')

ax[1, 0].barh(catsales.index,catsales.sales,color="m")
ax[1,0].set_xlabel('sales')
ax[1,0].set_ylabel('category')
ax[1,0].set_title('category wise sales')
ax[1,1].barh(citysales.index,citysales.sales,color="g")
ax[1,1].set_ylabel('city')
ax[1,1].set_xlabel('sales')
ax[1,1].set_title('city wise sale')
ax[1,2].plot(ms.index,ms.sales,color="g",marker="o",ms=9,mec="r",mfc="k")
ax[1,2].set_ylabel('sales')
ax[1,2].set_xlabel('month num')
ax[1,2].set_title('month wise sales')

ax[2,0].plot(qs.index,qs.sales,color="g")
ax[2,0].set_ylabel('sales')
ax[2,0].set_xlabel('quarter')
ax[2,0].set_title('quarter wise sales')

ax[2,1].pie(snsales.sales,labels=snsales.index,textprops={"fontsize":8},autopct="%0.1f%%")
ax[2,1].set_title('shipping wise sales')

ax[2,2].barh(rqs.index,rqs.Quantity,color="y")
ax[2,2].set_ylabel('region')
ax[2,2].set_xlabel('quantity sold')

ax[2,2].set_title('region wise qty sold')
plt.show()
```
